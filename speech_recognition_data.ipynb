```python
import os
import random
# list file in folder
def list_files(folder):
    for r, d, f in os.walk(folder):
        for file in f:
            yield(os.path.join(r, file))
def count_files(folder):
    i=-1
    for i,file in enumerate(list_files(folder)):
        continue
    return i+1
def rand_gen_file(folder,n=1):
    file_num=count_files(folder)
    rand_num=random.sample(range(file_num), n)
    file_list=[]
    for i,file in enumerate(list_files(folder)):
        if i in rand_num:
            file_list.append(file)
        continue
    return file_list
    
```

# Step 2: Data augmentation
1. randomly add background noisy to 2000 random speech files to create new samples to data/speech_out folder
2. randomly merge two different background noises to create 2000 more samples in data/non_speech_out folder

```python
from pydub import AudioSegment
def sound_overlay(main_audio,background_audio,output_file):
    rand_num=random.sample(range(-10,-5), 1)
    sound1 = AudioSegment.from_wav(main_audio)
    sound2 = AudioSegment.from_wav(background_audio)
    combined_sounds = sound1.overlay(sound2+rand_num[0])
    combined_sounds.export(output_file, format="wav")
```

```python
#generate new speech files in data/new_speech for inspection
import os
if not os.path.exists("data/new_speech"):
    os.mkdir("data/new_speech")
rand_speech_flist=rand_gen_file(final_speech_folder,2000)
rand_background_flist=rand_gen_file(final_non_speech_folder,2000)
for i in range(2000):
    outfile=os.path.join("data","new_speech","speech"+str(i)+".wav")
    sound_overlay(rand_speech_flist[i],rand_background_flist[i],outfile)
```

```python
#generate new non_speech files in data/new_non_speech for inspection
os.mkdir("data/new_non_speech")
rand_background_flist_1=rand_gen_file(final_non_speech_folder,2000)
rand_background_flist_2=rand_gen_file(final_non_speech_folder,2000)
for i in range(2000):
    outfile=os.path.join("data","new_non_speech","non_speech"+str(i)+".wav")
    sound_overlay(rand_background_flist_1[i],rand_background_flist_2[i],outfile)
```

## change wave framerate to 16kHz

```python
from pydub import AudioSegment
def convert_framrate(folder):
    file_list=list_files(folder)
    for file in file_list:
        sound=AudioSegment.from_wav(file)
        sound = sound.set_frame_rate(16000)
        sound.export(file,format="wav")
convert_framrate(final_speech_folder)
convert_framrate(final_non_speech_folder)
```
